### Préparation des données et calculs

```python
# rappel du dataframe
data=pd.read_csv('data/p_notes.csv', index_col=0)
data.drop(['is_genuine'], axis=1, inplace = True)
data.head(1)
```

```python
# préparation des données pour l'ACP
X = data.values
names = data.index
features = data.columns

#transformation–centrage-réduction
X_scaled = preprocessing.StandardScaler().fit_transform(X)
```

```python
# df données centrées-réduites
data_cr = pd.DataFrame(X_scaled, columns = ['length', 'diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up']) 
data_cr.index.name='id'
data_cr.to_csv('data/notes_cr.csv')
data_cr.head(1)
```

```python
# nombre de composantes
n_comp = 6

# Calcul des composantes principales
pca = decomposition.PCA(svd_solver='full', n_components=n_comp)
pca.fit(X_scaled)
```

### Scree plot et variance

```python
def display_scree_plot(pca):
    sns.set_style("whitegrid")
    plt.figure(figsize=(8,6))
    
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlim(0.5, 6.5)
    plt.ylim(0, 102)
    plt.xlabel("rang de l'axe d'inertie", size=13, fontweight='bold')
    plt.ylabel("pourcentage d'inertie", size=13, fontweight='bold')
    plt.title("Eboulis des valeurs propres", color='navy', size=16)
    plt.axvline(2, linestyle='dashed', color = 'grey')
    plt.show(block=False)
```

```python
# Eboulis des valeurs propres
display_scree_plot(pca)

# plt.savefig('img/P6.1_scree_plot.png', dpi=120, bbox_inches='tight')
plt.show()
plt.close()  
```

```python
# Proportion de variance expliquée
ve=pca.explained_variance_ratio_ *100
# Proportion de variance cumulée
vc=np.cumsum(pca.explained_variance_ratio_)*100

var = pd.DataFrame(data=[ve, vc],index =['Variance expliquée', 'Variance cumulée'])
var.columns=['CP1','CP2','CP3','CP4','CP5','CP6']
var
```

**Analyse**<br>
* Le coude apparaît dès la deuxième composante.
* Le premier plan factoriel CP1-CP2 représente 69,4 % variance cumulée

```python
# df 'notes' variables centrées-réduites
X_scaled=pd.read_csv('data/notes_cr.csv', index_col =0)
X_scaled.head(1)
```

### k-means sur les données centrées-réduites

```python
#k-means sur les données centrées et réduites
from sklearn import cluster
kmeans = cluster.KMeans(n_clusters=2)
kmeans.fit(X_scaled)
```

```python
# cluster prédit pour chaque variable
identified_clusters = kmeans.fit_predict(X_scaled)
identified_clusters
```

```python
# distances aux centres de classes des observations
k2=kmeans.transform(X_scaled)
k2
```

```python
# pandas DataFrame de clusters
pkc = pd.DataFrame(identified_clusters, columns = ['clust']) 
pkc.index.name='id'

pkc['clust'] = pkc['clust'].astype('bool')

pkc.head(2)
```

### Projection des individus sur CP1-CP2

```python
# cluster auxquels appartiennent les billets
statut = [pkc.loc[statut, 'clust'] for statut in data.index]

# projection des individus
X_projected = pca.transform(X_scaled)
display_factorial_planes(X_projected, n_comp, pca, [(0,1)],  
                         illustrative_var = statut, alpha = 0.8)

#plt.savefig('img/P6.2_proj_indiv.png', dpi=120, bbox_inches='tight')
plt.show()
plt.close()  
```

**Analyse**<br>
* L’apprentissage non-supervisé K-means (avec K=2 clusters) est une méthode qui offre rapidement une solution - pas nécessairement optimale, par son appartenance à une classe de problèmes donnés déjà identifiés (heuristique)
* Notre modèle, projeté sur les plans factoriels 1 et 2 propose une solution de classement proche de celle fournie par le statut réel des individus (is_genuine) sur les composantes 1 et 2

```python
#librairie pour évaluation des partitions
from sklearn import metrics

#utilisation de la métrique "silhouette"
#faire varier le nombre de clusters de 2 à 10
res = np.arange(6,dtype="double")
for k in np.arange(6):
    km = cluster.KMeans(n_clusters=k+2)
    km.fit(X_scaled)
    res[k] = metrics.silhouette_score(X_scaled,km.labels_)
print(res)
```
